```python
# 파일 불러오기

import pymupdf4llm

first_file = "C:/Users/user/Desktop/BOAZ_ADV/Wang_Gyu/1-s2.0-S0952818016300204-main.pdf"
second_file = "C:/Users/user/Desktop/BOAZ_ADV/Wang_Gyu/1-s2.0-S0952818022000101-main.pdf"
```

```python
# pymupdf 수행

first_pdf = pymupdf4llm.to_markdown(first_file,
                                  page_chunks=True,
                                  write_images=True,
                                    image_path="first_images",
                                    image_format="png")

second_pdf = pymupdf4llm.to_markdown(second_file,
                                  page_chunks=True,
                                  write_images=True,
                                    image_path="second_images",
                                    image_format="png")
```

```python
# pdf 페이지 확인
print(f"총 {len(first_pdf)}페이지")
print('--------------------')

# 가져온 정보 확인
print(f"주요 키 : {first_pdf[0].keys()}")
print('--------------------')

# metadata
print(first_pdf[0]['metadata'])
print('--------------------')

# toc_items -> 각 페이지 제목/소제목 
print(first_pdf[0]['toc_items'])
print('--------------------')

# tables
print(first_pdf[0]['tables'])
print('--------------------')

# images
print(first_pdf[0]['images'])
print('--------------------')

# graphics
print(first_pdf[0]['graphics'])
print('--------------------')

# text
print(first_pdf[0]['text'])
print('--------------------')
```

```python
# first pdf 첫 번째 페이지가 문단 순서 안 맞음

# \n\n 기준으로 나누기
text = first_pdf[0]['text']
split_text = text.split('\n\n')

# 문장 순서 맞추기
split_text[8], split_text[9] = split_text[9], split_text[8]

split_text[10], split_text[11] = split_text[11], split_text[10]

split_text
```

```python
# 문장 합치기
split_text[2] = split_text[2] + '\n' + split_text[3] + '\n' + split_text[4] + '\n' + split_text[5] + '\n' + split_text[6] 
split_text[9] = split_text[9] + '\n' + split_text[10]
split_text[13] = split_text[13] + '\n' + split_text[14] + '\n' + split_text[15]

split_text
```

```python
del split_text[16]

# 합친 문장 지우기
del split_text[15]
del split_text[14]
del split_text[10]
del split_text[6]
del split_text[5]
del split_text[4]
del split_text[3]

split_text
```

```python
# 마크 다운 형식으로 변하지 않은 놈들 바꾸기
# ex) Abstract 강조하기

import re

split_text = [re.sub(r'\bAbstract\b', '**Abstract:**', line) for line in split_text]
split_text = [re.sub(r'\bKeywords:\b', '**Keywords:**', line) for line in split_text]
split_text = [re.sub(r'\b1\. Introduction\b', '### 1. Introduction', line) for line in split_text]

# 첫 페이지 최종
split_text
```

```python
# pdf 첫 페이지 빼고 \n\n 기준으로 다 합치기

total_text = []
for i in range(len(first_pdf)-1):
    pdf_text = first_pdf[i+1]['text'].split('\n\n')
    for j in pdf_text:
        total_text.append(j)

total_text
```
